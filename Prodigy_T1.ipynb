```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the insurance dataset
insurance_df = pd.read_csv('insurance.csv')

# Plot the distribution of ages
plt.figure(figsize=(10, 6))
plt.hist(insurance_df['age'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Ages in the Insurance Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the insurance dataset
insurance_df = pd.read_csv('insurance.csv')

# Count the occurrences of each gender
gender_counts = insurance_df['smoker'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
gender_counts.plot(kind='bar', color=['skyblue', 'pink'], edgecolor='black')
plt.title('Distribution of Smokers in the Insurance Dataset')
plt.xlabel('Smoker')
plt.ylabel('Frequency')
plt.xticks(rotation=0) # Rotate x-axis labels if needed
plt.grid(axis='y') # Add gridlines on the y-axis
plt.show()
```
